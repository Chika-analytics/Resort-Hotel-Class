```python
class ResortRoom:
    
    def __init__(self, room_number):
        self.room_number = room_number
        self.guest_name = None
        self.is_booked = False

    def book_room(self, guest_name):
        
        if self.is_booked:
            print("Room already booked!")
        else:
            self.guest_name = guest_name
            self.is_booked = True
            print(f"Room {self.room_number} booked for {guest_name}")

    def checkout(self):
        
        if self.is_booked:
            print(f"{self.guest_name} checked out")
            self.guest_name = None
            self.is_booked = False
        else:
            print("Room is already empty")

    def display_info(self):
        print("Room Number:", self.room_number)
        print("Guest:", self.guest_name)
        print("Booked:", self.is_booked)

room1 = ResortRoom(101)
room1.book_room("Alice")
room1.display_info()
```

```python
room2 = ResortRoom(102)
room2.book_room("David")
```

```python
room1.checkout()
room1.display_info()
```

```python
room2.checkout()
room2.display_info()
```
